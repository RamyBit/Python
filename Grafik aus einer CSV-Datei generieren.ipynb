```python
xs =[]
ys = []

name = "Anna"
gender ="F"
state = "CA"

with open("data/names.csv", "r") as file:
    for line in file:
        data = line.strip().split(",")
        if data[1] == name and data[3] == gender and data[4] == state:
            xs.append(data[2])
            ys.append(data[5])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.plot(xs, ys)
plt.show()
```
